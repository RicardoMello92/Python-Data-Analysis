Classificação dos livros do acervo com base nas Classes abaixo

000 a 099: Generalidades. Ciência e conhecimento.

100 a 199: Filosofia e psicologia.

200 a 299: Religião.

300 a 399: Ciências sociais.

400 a 499: Classe vaga. Provisoriamente não ocupada.

500 a 599: Matemática e ciências naturais.

600 a 699: Ciências aplicadas.

700 a 799: Belas artes.

900 a 999: Geografia. Biografia. História.


```python
class_list=[]
for i in dados['localizacao']:
  if i in range (0,100):
    class_list.append('Generalidades. Ciência e conhecimento')
  elif i in range (101,200):
    class_list.append('Filosofia e psicologia')
  elif i in range (201,300):
    class_list.append('Religião')
  elif i in range (301,400):
    class_list.append('Ciências sociais')
  elif i in range (401,500):
    class_list.append('Classe vaga (Provisoriamente não ocupada)')
  elif i in range (501,600):
    class_list.append('Matemática e ciências naturais')
  elif i in range (601,700):
    class_list.append('Ciências aplicadas')
  elif i in range (701,800):
    class_list.append('Belas artes')
  elif i in range (801,900):
    class_list.append('Linguagem. Lingua. Lingúistica')
  else:
    class_list.append('Geografia. Biografia. História')

```
